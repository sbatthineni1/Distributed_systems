```python
def read_from_text_file():
    with open("inp.txt",'r') as f:
        return int(f.read())
```

```python
def process():
    with open('inp.txt', "r") as fptr:
        N = read_from_text_file()
        print("Contents of the file is",N)
        pid = os.getpid()
        print("N: %d Process ID: %d",N,pid)
        
    with open('inp.txt', "w") as fptr:
        N+=1
    
        write_to_txt_file(N)
        fptr.flush()
```

```python
def create_txt_file():
    n = input("Please enter a number for the file")
    write_to_txt_file(n)
```

```python
def write_to_txt_file(inp):
    inp = str(inp)
    with open("inp.txt",'w') as f:
        f.write(inp)
```
